```python
Types = {
    'aisle_id': 'int32',
    'aisle': 'category',
    'order_id': 'int32',
    'user_id': 'int32',
    'eval_set': 'category',
    'order_number': 'int32',
    'order_dow': 'int32',
    'order_hour_of_day': 'int32',
    'days_since_prior_order': 'float16',
    'department_id': 'int8',
    'department': 'category',
    'product_id': 'int32',
    'product_name': 'category',
    'add_to_cart_order': 'int32',
    'reordered': 'int8',
}


# Reading files using enhanced data types
aisles = pd.read_csv(data + 'aisles.csv', dtype={i: Types.get(i, None) for i in ['aisle_id','aisle']})

departments = pd.read_csv(data + 'departments.csv', dtype={i: Types.get(i, None) for i in ['department_id','department']})

products = pd.read_csv(data + 'products.csv', dtype={i: Types.get(i, None) for i in ['product_id', 'aisle_id', 'department_id',' product_name']})

# Read orders, excluding 'days_since_prior_order' from explicit dtype setting during read_csv
orders = pd.read_csv(data + 'orders.csv', dtype={i: Types.get(i, None) for i in ['order_id', 'user_id', 'order_number', 'order_dow', 'order_hour_of_day','eval_set']})
# Convert 'days_since_prior_order' to float16 after loading the DataFrame
orders['days_since_prior_order'] = orders['days_since_prior_order'].astype('float16')

order_products_prior = pd.read_csv(data + 'order_products__prior.csv', dtype={i: Types.get(i, None) for i in ['order_id', 'product_id', 'add_to_cart_order', 'reordered']})
```

```python
print(aisles.info())
print("------------------------------------------------")
print(orders.info())
print("------------------------------------------------")
print(departments.info())
print("------------------------------------------------")
print(products.info())
print("------------------------------------------------")
print(order_products_prior.info())
```

```python
My_Data = pd.merge(orders, order_products_prior,how = 'left', on='order_id')
My_Data = pd.merge(My_Data, products,how = 'left', on='product_id')
My_Data = pd.merge(My_Data, aisles,how = 'left', on='aisle_id')
My_Data = pd.merge(My_Data, departments,how = 'left', on='department_id')
My_Data.head()
```

#**Cleaning and Handling Missing Values**

### Median Imputation

```python
median_df = orders.copy()
median_value = median_df['days_since_prior_order'].median()
median_df['days_since_prior_order_median'] = median_df['days_since_prior_order'].fillna(median_value)
print(f"Median value used: {median_value}")
print(f"Missing values after median imputation: {median_df['days_since_prior_order_median'].isna().sum()}")
```

### Mode Imputation

```python
mode_df = orders.copy()
mode_value = mode_df['days_since_prior_order'].mode()[0]
mode_df['days_since_prior_order_mode'] = mode_df['days_since_prior_order'].fillna(mode_value)
print(f"Mode value used: {mode_value}")
print(f"Missing values after mode imputation: {mode_df['days_since_prior_order_mode'].isna().sum()}")
```

### Mean Imputation

```python
mean_df = orders.copy()
mean_value = mode_df['days_since_prior_order'].mean()
mean_df['days_since_prior_order_mode'] = mean_df['days_since_prior_order'].fillna(mean_value)
print(f"Mode value used: {mean_value}")
print(f"Missing values after mode imputation: {mean_df['days_since_prior_order_mode'].isna().sum()}")
```

### Sentinel Value Imputation

```python
sentinel_df = orders.copy()
sentinel_df['days_since_prior_order_sentinel'] = sentinel_df['days_since_prior_order'].fillna(-1)
print(f"Sentinel value used: -1")
print(f"Missing values after sentinel imputation: {sentinel_df['days_since_prior_order_sentinel'].isna().sum()}")
```

###Outlier detection & treatment : Winsorizing

```python
order_products = pd.concat([order_products_prior])
basket_size = order_products.groupby('order_id').size()
orders['basket_size'] = orders['order_id'].map(basket_size).fillna(0).astype('int16')

upper = orders['basket_size'].quantile(0.99)
print(f"99th percentile: {upper} items")
print(f"Number of orders > {upper}: {(orders['basket_size'] > upper).sum()}")
print(f"Percentage: {(orders['basket_size'] > upper).mean()*100:.3f}%")

orders['basket_size_capped'] = orders['basket_size'].clip(lower=orders['basket_size'].quantile(0.01),
  upper=orders['basket_size'].quantile(0.99))

print(f"Max after capping: {orders['basket_size_capped'].max()}")
```
